```python
import numpy as np
import matplotlib.pyplot as plt

test_1 = np.array([5, 3, 7, 9, 8, 3])
test_2 = np.array([5, 8, 8, 2, 7, 6])
test_3 = np.array([10, 6, 4, 3, 5, 4])
x = np.arange(len(test_1))
students = ["Alex", "Katie", "Maya", "Felix", "Lauren", "Arjun"]
fig, axs = plt.subplots(2, 2, figsize=(12,10))

# Bar chart
axs[0,0].bar(students, test_1)
axs[0,0].set_title('Bar Graph')

# Line graph
axs[0,1].plot(x, test_1, label="Test 1", marker='*', linestyle=":")
axs[0,1].plot(x, test_2, label="Test 2", marker='o', linestyle="-.")
axs[0,1].set_title('Line Graph')
axs[0,1].legend()

# Scatter plot
axs[1,0].grid()
axs[1,0].scatter(test_1, test_2, s=60)
axs[1,0].set_title('Scatter Plot')
axs[1,0].set_xlabel('Test 1 score')
axs[1,0].set_ylabel('Test 2 score')

# Stacked bar chart
axs[1,1].bar(students, test_1, label="Test 1", color='indigo')
axs[1,1].bar(students, test_2, label="Test 2", bottom=test_1, color="violet")
axs[1,1].bar(students, test_3, label="Test 3", bottom=test_1+test_2, color="purple")
axs[1,1].legend()
axs[1,1].grid(axis='y')
axs[1,1].set_axisbelow(True)
axs[1,1].set_title('Vertical Stacked Bar Chart')
axs[1,1].set_xlabel('Total test scores');
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(12,10))

x = np.linspace(-np.pi, 4 * np.pi, 1860)
y = np.sin(x)
z = np.cos(x)

axs[0,0].plot(x,y,label="Sin")
axs[0,0].plot(x,z, label="Cos")
axs[0,0].legend()
```
